## Part 2 - Univariate Analysis and Box Plot I
I wish to investigate the number of substitutions going in the field based on positions. I will be using a boxplot to display my findings because it allows me to determine a quick graphical view. I am curious to find out the position with the highest and lowest amount of substitusions or any suprise positions.

```python
#Using figsize (x,y) to determine the size of the graph
plt.figure(figsize=(13, 6))
#Applying box plots using seaborn
sns.boxplot(data=df, x='Position', y='Number Of Substitute In', color="#FF8800")
#Setting the title, xlabel and ylabel using plt
plt.title('Number of Substitutes In based on Position')
plt.xlabel('Position')
plt.ylabel('Number Of Substitute In')
plt.xticks(rotation=45)
plt.grid(True)
```

After plotting the graph we can extrapolate some points from it. We can see the Attacking Midfielder position experiences the highest frequency of substitutions, with the median number of substitutions being 12 and an interquartile range of 4. This could mean that teams tend to rotate players in this position more frequently, possibly due to tactical adjustments or the demanding nature of the role. Similarly the Right Winger position also experiences the median and interquartile range which is suprising, on top of that we can observe that it also has the highest outlier of 13. As expected we can we see that the goalkeeper positions stands out as the only one without any substitution, as goalkeepers typically play the entire duration of the match due to their stationary role. I have also included the substitute out below for comparison.

```python
#Using figsize (x,y) to determine the size of the graph
plt.figure(figsize=(13, 6))
#Applying box plots using seaborn
sns.boxplot(data=df, x='Position', y='Number Of Substitute Out', color="#fb6f92")
#Setting the title, xlabel and ylabel using plt
plt.title('Number of Substitutes Out based on Position')
plt.xlabel('Position')
plt.ylabel('Number Of Substitute Out')
plt.xticks(rotation=45)
plt.grid(True)
```
