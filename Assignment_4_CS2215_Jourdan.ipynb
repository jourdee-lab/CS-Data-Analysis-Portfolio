## Part 2 - Scatter Plot

I want to investigate if the Goals and Market really have an affect on the age of players. I will use a scatter plot to display my findings as it allows me to see if there's a pattern at play. I wish to understand if players do really get a higher market valuation based the number of scored goals and age of the players and see if there's any outliers in the mix.

```python
df.plot.scatter(x="Markey Value In Millions(£)",y="Goals",c="Age",
                          marker="^",
                          title= "Goal and Market Value relative to the Age of players",
                          xlabel="Market Value of Players",
                          ylabel="Goals of Players",
                          figsize=(16, 6),
                          alpha=.8,
                          s=32)
```

Looking at the graph above we see a cluster of data points in the bottom left region, suggesting that the majority of players have a market value between 20 and 80 years of age and have scored between 0-10 goals on average. As indicated by the colour coding on the right axis representing Age, there are a few outliers. Notably, one significant outlier is visible in the top area with the highest goal total of 24 and with a market value of 55 million, their Age falls above the higher range of 35. This means that this one outlier is performing at a high level scoring lots of goal with a modest market value while being at the older range of players. When considering the full data set depicted, there are also several instances of players getting higher market evaluation despite a younger ages and lower goal totals. This indicates that goals scored and market value may not be as strongly correlated to a player's age as initially thought. This shows that goals and Market value does not impact the players age as much as I thought it would.  

## Part 3 - Area Plot

I want to investigate the growth and decline of google stock with the open, low, high and close prices over a period of time. I will display it using an area plot to give us an overview of the volatility of the Google's stock. I wish to understand when and where the periods of fluctuations are, to understand why it happened.


```python

df2.plot.area(x="Date", y=["Open","Low","High","Close","Adj Close"],
             alpha=0.7,
             figsize=(17, 8),
             grid = True,
             xlabel="Years",
             ylabel="Stock Price in ($)",
             title= "Google Stock Price Trends overtime").legend(loc=(1.01, 0.825))
```

According to the graph above we can extract a few major points from it. During the start of the 2004 year we see that google's stock price start at a low 50 dollars per stock and gradually increased over the year with a sudden peak at 2007 of 300. However, we can see that in 2008 stock price fell gradually due the 2008 market crash, but in general after that point google's market volatility exhibit a consistent upward trend with a few occasioanal periods of volatity. We also the areas for "High" and "Low" prices are generally wider than the "Close" price, indicating periods of high volatility. This could mean that Google's stock experiences significant fluctuations in price during at those time periods. One major dip happened in the early periods of 2020 which is during the start of the pandemic, we see a really steep decline this might be because of the reduced number of trades which resulted in a lower stock price. I have also included the line plot below for a more a easier to read representation of the price over time.

```python
df2.plot.line(x="Date", y=["Open","Low","High","Close","Adj Close"],
             alpha=0.7,
             figsize=(17, 8),
             grid = True,
             xlabel="Years",
             ylabel="Stock Price in ($)",
             title= "Google Stock Price Trends overtime").legend(loc=(1.01, 0.825))
```
